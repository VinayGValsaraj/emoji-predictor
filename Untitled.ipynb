```python
import json
import os
from emoji import UNICODE_EMOJI

with open('data/29.json', 'r') as file:
    contents = file.readlines()
    tweets = []
    for line in contents:
        tweets.append(json.loads(line))
#     print (len(tweets))
```

```python
print(len(tweets))
```

```python
count = 0
emoji_tweets = []
for idx,tweet in enumerate(tweets):
    if 'lang' in tweet:
        if tweet['lang'] == 'en':
            count += 1
            for word in tweet['text']:
                if word in UNICODE_EMOJI:
                    print(idx, tweet['text'])
                    emoji_tweets.append(tweet['text'])
                    break
        

        
```
