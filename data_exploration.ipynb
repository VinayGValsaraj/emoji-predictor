```python
from utils import get_train_data_from_csv, get_indexer
from random import shuffle
```

```python
data = get_train_data_from_csv('data/train_cont.csv')
```

```python
indexer = get_indexer('indexer_cont.csv')
```

```python
def check_done(labels):
    all_5 = True
    
    if len(labels) != 10:
        all_5 = False
    for key in labels:
        if len(labels[key]) != 5:
            all_5 = False
            
    return all_5
        
```

```python
labels = {}
for d in data:
    if labels.get(d.label) is None:
        labels[d.label] = [d.text]
    else:
        if len(labels[d.label]) != 5:
            labels[d.label].append(d.text)
        
    if check_done(labels):
        break
    
```

```python
for label in labels:
    for text in labels[label]:
        print(indexer.get_object(label), ": ", text)
    print()
```
